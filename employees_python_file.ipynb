```python
df1 = pd.DataFrame({'stud_id': ['101', '102', '103', '104'],
                     'exam_no1': ['800', '900', '750', '800'],
                     'exam_no2': ['700', '850', '650', '900']})
   
df2 = pd.DataFrame({'stud_id': ['101', '107', '103', '104'],
                          'exam_no1': ['800', '600', '720', '650'],
                          'exam_no2': ['650', '700', '900', '900']}) 
```
